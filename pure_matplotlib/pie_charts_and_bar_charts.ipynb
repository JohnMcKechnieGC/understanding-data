# Pie Charts and Bar Charts

## Scenario
Imagine you manage a small cafe and track daily sales for various product categories. For example, let's say you want to visualize:
1. The proportion of total sales contributed by different product categories (e.g., Coffee, Tea, Pastries, Sandwiches).
2. The number of units sold for each product category in a given week.

We'll then create:
- A pie chart to show the percentage of total sales each category represents.
- A bar chart to compare the sales across categories by quantity sold.

```python
import matplotlib.pyplot as plt
```

For this example, let's say the cafe sells four categories of items. We'll define the data as follows:
- Categories: Coffee, Tea, Pastries, Sandwiches
- Daily revenue (Â£) made from each category
- Units sold per category over a week

Next, we'll create some test data:

```python
# Categories
categories = ["Coffee", "Tea", "Pastries", "Sandwiches"]

# Example daily revenue in pounds
daily_revenue = [350, 150, 100, 200]

# Example weekly units sold (over 7 days)
# Let's say we recorded total items sold for each category in one week
units_sold = [700, 350, 200, 400]
```

## Pie Chart

```python
# Create a new figure for the pie chart
plt.figure(figsize=(6, 6))

# Create a pie chart
#plt.pie(daily_revenue)

plt.pie(daily_revenue,
        labels=categories,
        autopct='%1.1f%%',
        startangle=90
        )

# Add a title
plt.title("Proportion of Daily Revenue by Category")

# Display the pie chart
plt.show()
```

### 'Exploding' pie charts

```python
# Create a new figure for the pie chart
plt.figure(figsize=(6, 6))

# Create a pie chart
# plt.pie(daily_revenue)

plt.pie(daily_revenue,
        labels=categories,
        autopct='%1.1f%%',
        startangle=90,
        explode=[0, 0.1, 0, 0.2]
        )

# Add a title
plt.title("Proportion of Daily Revenue by Category")

# Display the pie chart
plt.show()
```

## Bar Chart

```python
# Create a new figure for the bar chart
plt.figure(figsize=(8, 5))

# Create a bar chart
plt.bar(categories,
        units_sold,
        color='green'
        )

# Add a title and labels
plt.title("Weekly Units Sold by Category")
plt.xlabel("Product Category")
plt.ylabel("Units Sold (per week)")

# Add a grid for better readability
plt.grid(axis='y',
        linestyle='--',
        alpha=0.7
        )

# Display the bar chart
plt.show()
```

# Customizing the Colour Palette
This section redraws the pie chart using a custom colour palette.

```python
# Define a custom colour palette for the pie chart
# We'll choose a selection of complementary colours for each category using hex values
custom_colours = ["#8B4513",  # Coffee: saddle brown
                  "#DAA520",  # Tea: goldenrod
                  "#CD853F",  # Pastries: peru
                  "#DEB887"]  # Sandwiches: burlywood

# Here we show how to do the same thing but this time using RGB values
# custom_colours = [
#     (0.545, 0.270, 0.0745),  # Coffee
#     (0.855, 0.647, 0.125),  # Tea
#     (0.8039, 0.5216, 0.2471),  # Pastries
#     (0.8706, 0.7216, 0.5294)  # Sandwiches
# ]

# Set our figure size.
plt.figure(figsize=(6, 6))

# Create a pie chart with custom colours
plt.pie(daily_revenue,
        labels=categories,
        autopct='%1.1f%%',
        startangle=90,
        colors=custom_colours)

# Add a title
plt.title("Proportion of Daily Revenue by Category")

# Display the pie chart
plt.show()
```

And here we show how to redraw the bar chart with the custom colour palette.

```python
# Set the figure size
plt.figure(figsize=(8, 5))

# Create a bar chart with individual colours per bar
plt.bar(categories, units_sold, color=custom_colours)

# Add a title and labels
plt.title("Weekly Units Sold by Category")
plt.xlabel("Product Category")
plt.ylabel("Units Sold (per week)")

# Add a grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
